```python
import pandas as pd
import dataframe_image as dfi
import json
import os

path = './' 
versions = ['v1', 'v2', 'v3', 'v4']
data_list = []

for v in versions:
    file_name = f'summary_{v}.json'
    file_path = os.path.join(path, file_name)
    
    if os.path.exists(file_path):
        with open(file_path, 'r') as f:
            try:
                data = json.load(f)
                data['Version'] = v
                data_list.append(data)
            except json.JSONDecodeError:
                print(f"Warning: {file_name} wrong format, skipped")


if data_list:
    df = pd.DataFrame(data_list)

    metrics = [c for c in df.columns if c != 'Version']
    df = df[['Version'] + metrics]

    styled_df = df.style.set_caption("RAG Metrics Comparison") \
    .background_gradient(cmap='Blues', subset=metrics)

    dfi.export(styled_df, 'rag_metrics_comparison.png', table_conversion='matplotlib')

    styled_df
```
